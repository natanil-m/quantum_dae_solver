```python
import numpy as np
from scipy.optimize import minimize


def polynomial(a, b, c, d, t):
    return a * t**3 + b * t**2 + c * t + d

t_data = np.linspace(0, 10, 100)
target_values = 2 * t_data**3 + 3 * t_data**2 + 4.5 * t_data + 2


def loss(params):
    a,b,c,d = params
    fitted_values = polynomial(a, b, c, d,t_data)
    return np.sum((fitted_values - target_values) ** 2)


initial_params = [1, 1, 1,1]


result = minimize(loss, initial_params, method='Nelder-Mead')

a, b, c,d = result.x


print("Optimized a:", a)
print("Optimized b:", b)
print("Optimized c:", c)
print("Optimized d:", d)
```

```python
import matplotlib.pyplot as plt

# Plot the target values and the fitted curve
plt.figure(figsize=(8, 6))
plt.scatter(t_data, target_values, label='Target Values', alpha=0.5)
fitted_values = polynomial(a, b, c, d,t_data)
plt.plot(t_data, fitted_values, 'r', label='Fitted Curve')
plt.xlabel('Time (t)')
plt.ylabel('Values')
plt.legend()
plt.title('Fitting Data')
plt.show()
```
