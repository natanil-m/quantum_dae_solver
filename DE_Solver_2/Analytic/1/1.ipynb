```python
import numpy as np
import matplotlib.pyplot as plt

def delta(t):
    sqrt_10 = np.sqrt(10)
    result = (7 * np.sin(sqrt_10 * t) / sqrt_10) - (3 * np.cos(sqrt_10 * t) / 2) + 1/2
    return result

# Generate a range of t values
t_values = np.linspace(0, 1, 50)  # Adjust the range and number of points as needed

# Compute the corresponding delta values
delta_values = [delta(t) for t in t_values]

# Plot delta(t)
plt.figure(figsize=(8, 6))
plt.plot(t_values, delta_values, label='delta(t)')
plt.xlabel('t')
plt.ylabel('delta(t)')
plt.title('Plot of delta(t)')
plt.grid(True)
plt.legend()
plt.show()
```
