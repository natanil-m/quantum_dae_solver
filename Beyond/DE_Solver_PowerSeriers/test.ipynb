```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the parameters
m = 1.0  # Mass
b = 0.2  # Damping coefficient
k = 1.0  # Spring constant

# Define the input function u(t) (you'll need to specify this function)
def u(t):
    # Define your input function here. For example, a step input:
    return 1.0 if t >= 0 else 0.0

# Define the ODE as a system of first-order ODEs
def system(y, t):
    y1, y2 = y[0], y[1]
    y1_dot = y2
    y2_dot = (u(t) - b * y2 - k * y1) / m
    return [y1_dot, y2_dot]

# Initial conditions
y0 = [0.0, 0.0]

# Time points for integration
t = np.linspace(0, 20, 1000)  # Adjust the time span as needed

# Solve the ODE system
sol = odeint(system, y0, t)

# Extract the solution for y(t)
y = sol[:, 0]

# Plot the solution
plt.figure()
plt.plot(t, y)
plt.xlabel('Time')
plt.ylabel('y(t)')
plt.title('Solution of the Second-Order ODE')
plt.grid(True)
plt.show()
```
