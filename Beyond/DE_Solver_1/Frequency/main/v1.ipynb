```python
def f(t,freqs,coef,a0):
    # print(t)
    # print(freqs)
    # print(coef)
    # print(np.sin(freqs*t))
    in_sigma = qnp.array(coef*(qnp.e**(-t)*qnp.sin(freqs*t)))
    # print(np.sum(in_sigma))
    return qnp.sum(in_sigma)+a0
```
