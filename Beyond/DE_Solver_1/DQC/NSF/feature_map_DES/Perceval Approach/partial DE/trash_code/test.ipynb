```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range of x and t values
x = np.linspace(0, 1, 10)  # Range of x values
t = np.linspace(0, 1, 10)  # Range of t values

# Create a meshgrid of x and t values
X, T = np.meshgrid(x, t)

# Compute the values of U(x, t)
U = X**2 + T**2 + X*T

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, T, U, cmap='viridis')

# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('U(x, t)')
ax.set_title('Plot of U(x, t) = x^2 + t^2 + xt')

# Show the plot
plt.show()
```
