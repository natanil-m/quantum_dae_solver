```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the differential equation
def equation(x, f):
    return [f[1], -np.sin(x)]

# Set the initial conditions
x0 = 0.0  # Initial value for x
f0 = [0.0, 1.0]  # Initial values for f(x) and its derivative

# Define the range of x values to solve for
x_span = [0.0, 1.0]  # Start and end values of x

# Solve the differential equation
solution = solve_ivp(equation, x_span, f0,t_eval=np.linspace(x_span[0], x_span[1], 100))

# Extract the solution
x_values = solution.t
f_values = solution.y[0]

# Plot the solution
plt.plot(x_values, f_values)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Solution of f\'\'(x) + sin(x) = 0')
plt.grid(True)
plt.show()
```
